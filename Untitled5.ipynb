```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
fear_greed = pd.read_excel('fear_greed_index.xlsx')
trades = pd.read_csv('historical_data.csv')
```

```python
print(fear_greed.head())
print(trades.head())
```

```python
print(fear_greed.columns)
print(trades.columns)
```

```python
fear_greed['date'] = pd.to_datetime(fear_greed['date'])

trades['Timestamp IST'] = pd.to_datetime(
    trades['Timestamp IST'],
    format='%d-%m-%Y %H:%M'
)

trades['date'] = trades['Timestamp IST'].dt.date

trades['date'] = pd.to_datetime(trades['date'])
```

```python
merged = pd.merge(
    trades,
    fear_greed,
    on='date',
    how='left'
)

print(merged.head())
```

```python
print(merged.isnull().sum())
```

```python
merged.dropna(inplace=True)
```

```python
profit = merged.groupby('classification')['Closed PnL'].mean()

print(profit)
```

```python
profit.plot(kind='bar')
plt.title('Average Profit by Market Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Average Closed PnL')
plt.show()
```

```python
volume = merged.groupby('classification')['Size USD'].sum()
print(volume)
```

```python
volume.plot(kind='bar')
plt.title('Trading Volume by Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Trading Volume')
plt.show()
```

```python
merged['win'] = merged['Closed PnL'] > 0
win_rate = merged.groupby('classification')['win'].mean() * 100
print(win_rate)
```

```python
win_rate.plot(kind='bar')
plt.title('Win Rate by Sentiment')
plt.ylabel('Winning Trade Percentage')
plt.show()
```

```python
side_analysis = pd.crosstab(
    merged['classification'],
    merged['Side']
)
print(side_analysis)
```

```python
side_analysis.plot(kind='bar')
plt.title('Buy vs Sell Trades by Sentiment')
plt.ylabel('Number of Trades')
plt.show()
```

```python
corr = merged[['Closed PnL', 'Size USD']].corr()
print(corr)
```

```python
sns.heatmap(corr, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

The analysis shows that market sentiment significantly affects trader behavior. Greed periods resulted in higher trading volume and larger position sizes, while Fear periods reflected more cautious participation. Increased activity during Greed did not always improve win rates, suggesting emotional trading behavior and higher risk exposure.
